Question 3: What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?

->Autoencoders:
They learn to compress text into a hidden representation and then reconstruct the same text. They mainly focus on reconstruction, not generating new text.

Variational Autoencoders (VAEs):
They learn a probability distribution of the latent space, so they can generate new text samples, not just reconstruct the input.

Example: Variational Autoencoder

Key Difference:

Autoencoder: reconstructs the same input text.

VAE: learns a distribution and generates new text from it.

Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

->In Neural Machine Translation, the attention mechanism helps the model focus on important words of the input sentence while translating.

How it works:

The encoder processes the input sentence.

During translation, the decoder looks at different parts of the input sentence.

Attention gives higher weight to the most relevant words for generating the next word.

Example: Models like Transformer use attention for translation.

Why it is critical:

Handles long sentences better

Improves translation accuracy

Allows the model to focus on relevant words during translation.

```python
'''Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.'''

import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Attention
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

eng = ["hello", "how are you", "good morning", "thank you"]
spa = ["hola", "como estas", "buenos dias", "gracias"]

tok_eng = Tokenizer()
tok_spa = Tokenizer()

tok_eng.fit_on_texts(eng)
tok_spa.fit_on_texts(spa)

eng_seq = tok_eng.texts_to_sequences(eng)
spa_seq = tok_spa.texts_to_sequences(spa)

max_len = max(max(len(i) for i in eng_seq), max(len(i) for i in spa_seq))

eng_seq = pad_sequences(eng_seq, maxlen=max_len, padding="post")
spa_seq = pad_sequences(spa_seq, maxlen=max_len, padding="post")

vocab_eng = len(tok_eng.word_index) + 1
vocab_spa = len(tok_spa.word_index) + 1

encoder_inputs = Input(shape=(max_len,))
enc_emb = Embedding(vocab_eng, 64)(encoder_inputs)
encoder_lstm = LSTM(64, return_sequences=True)
encoder_outputs = encoder_lstm(enc_emb)

attention = Attention()
attention_output = attention([encoder_outputs, encoder_outputs])

decoder_lstm = LSTM(64, return_sequences=True)
decoder_outputs = decoder_lstm(attention_output)

outputs = Dense(vocab_spa, activation="softmax")(decoder_outputs)

model = tf.keras.Model(encoder_inputs, outputs)

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(eng_seq, np.expand_dims(spa_seq, -1), epochs=200)

test = tok_eng.texts_to_sequences(["hello"])
test = pad_sequences(test, maxlen=max_len, padding="post")

pred = model.predict(test)

pred_ids = np.argmax(pred[0], axis=1)

reverse_spa = {v:k for k,v in tok_spa.word_index.items()}

translated = " ".join([reverse_spa.get(i,"") for i in pred_ids if i!=0])

print("Input:", "hello")
print("Predicted Spanish:", translated)
```
